```python
# Quantidade de colunas com mais de 4930 observações não nulas
(Dados.notna().sum() >= 4930).sum()
```

É possível perceber que após as variáveis de idade média, todas as outras features apresentam menos que 4930 observações, sendo 21 a quantidade de colunas com mais do que 4930 observações não nulas

```python
# Retirar variáveis com menos que 4930 observações
def Retirar_colunas_Dados_Faltantes(Dados):
    colunas_obs_menores_4930 = Dados.columns[Dados.notna().sum() <= 4930]
    Dados_Sem_Colunas = Dados.drop(colunas_obs_menores_4930, axis="columns")
    return Dados_Sem_Colunas

Dados_Sem_Obv_Faltantes = Retirar_colunas_Dados_Faltantes(Dados)
Dados_Sem_Obv_Faltantes
```
